Image to pencil sketch conversion using opencv library of python. 

1.importing library and reading image.

```python

from google.colab.patches import cv2
old_image=cv2.imread("/content/Animals_Dogs_Cute_puppy_033686_.jpg")
old_image.shape
```

2. resizing the image.

```python
import cv2
image=cv2.resize(old_image,(520,520))
cv2_imshow(image)
cv2.waitKey(0)
```

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
cv2.waitKey(0)
```

```python
inverted_img=255- gray_img
cv2_imshow(inverted_img)
cv2.waitKey(0)
```

```python
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
cv2_imshow(blurred)
cv2.waitKey(0)
```

Result.

```python
from google.colab.patches import cv2_imshow
inverted_blurred=255 - blurred
pencil_sketch=cv2.divide(gray_img,inverted_blurred,scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
